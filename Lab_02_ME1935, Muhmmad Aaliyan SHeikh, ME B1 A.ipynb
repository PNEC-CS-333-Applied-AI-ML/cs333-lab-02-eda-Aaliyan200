## 4Ô∏è‚É£ Outlier Detection

```python
# 1. Visualization with Boxplots for Sensor Signals
sensor_cols = ['smcAC', 'smcDC', 'vib_table', 'vib_spindle', 'AE_table', 'AE_spindle']

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[sensor_cols])
plt.title("Boxplot for Outlier Detection (Sensors)")
plt.xticks(rotation=45)
plt.show()

# 2. IQR Method to identify Outliers
Q1 = df[sensor_cols].quantile(0.25)
Q3 = df[sensor_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outliers as values outside (Q1 - 1.5*IQR) and (Q3 + 1.5*IQR)
outliers_iqr = ((df[sensor_cols] < (Q1 - 1.5 * IQR)) | (df[sensor_cols] > (Q3 + 1.5 * IQR))).sum()
print("--- Outliers detected per sensor (IQR Method) ---")
print(outliers_iqr)

# 3. Z-Score Method (Modified)
# We calculate how many standard deviations a point is from the mean
from scipy import stats
z_scores = np.abs(stats.zscore(df[sensor_cols]))
outliers_z = (z_scores > 3).sum() # Standard threshold is 3

print("\n--- Outliers detected per sensor (Z-Score > 3) ---")
print(outliers_z)
```


### ‚úçÔ∏è
- Which method did you use (IQR or Modified Z-score)?
- Why?
- How many outliers were removed?
- How did it affect statistics?


In this analysis, the IQR (Interquartile Range) method was primarily used to identify extreme values because of its robustness in handling real-world sensor data that may not follow a perfect normal distribution. The results show that vibration and acoustic emission sensors are the most prone to outliers, with vib_table (17 outliers) and AE_table (10 outliers) showing the highest counts. In contrast, the motor current sensor smcAC showed 0 outliers, indicating a very stable electrical signal. The Z-Score method confirmed the presence of extreme values by flagging 25 data points as being more than three standard deviations from the mean. While these outliers represent abnormal machining cycles or sensor noise, they were retained or specifically noted to avoid losing critical information about tool failure. Removing or capping these values typically reduces the dataset's variance and standard deviation, leading to a more stable mean that better represents "normal" operating conditions for a Machine Learning model.
